```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.svm import LinearSVC
from sklearn.metrics import accuracy_score
```

Data preprocessing


```python
#load dataset 
raw_mail_data=pd.read_csv('/content/spam.csv',encoding='latin-1')
mail_data=raw_mail_data.where((pd.notnull(raw_mail_data)),'')
```

```python
mail_data.shape
```

```python
mail_data=mail_data.drop(mail_data.iloc[:,2:],axis=1)
```

```python
mail_data.columns=['Category','Message']
```

```python
mail_data.head()
```

```python
#label spam mail as 0 ,non spam mail as 1;
mail_data.loc[mail_data['Category']=='spam','Category',]=0
mail_data.loc[mail_data['Category']=='ham','Category',]=1
```

```python
#separate the data as text and label .X->text ,y_--->label
X=mail_data['Message']
y=mail_data['Category']
```

```python
print(X)
print('................')
print(y)
```

```python
#train test split
X_train,X_test,y_train,y_test=train_test_split(X,y,train_size=0.8,test_size=0.2,random_state=3)
```


feature extraction

```python
 #transform the text data to feature vectors that can be used as input to the svm model using TgidVectorizer
 #convert the text to lower case letters
 feature_extraction=TfidfVectorizer(min_df=1,stop_words='english',lowercase='True')
 x_train_features=feature_extraction.fit_transform(X_train)
 x_test_features=feature_extraction.transform(X_test) 

 #convert y_train and y_test values as integer
 y_train=y_train.astype('int')
 y_test=y_test.astype('int')
```

```python
feature_extraction.get_feature_names()
```

```python
x_train_features.shape
```

Support Vector Machine

```python
#training the svm with training data
model=LinearSVC()
model.fit(x_train_features,y_train)
```

EvaluAtion of the model

```python
#prediction on training data
prediction_on_training_data=model.predict(x_train_features)
accuracy_on_training_data=accuracy_score(y_train,prediction_on_training_data)
```

```python
print('accuracy on training data',accuracy_on_training_data)
```

```python
#prediction on test data
prediction_on_test_data=model.predict(x_test_features)
accuracy_on_test_data=accuracy_score(y_test,prediction_on_test_data)
```

```python
print('accuracy on test data',accuracy_on_test_data)
```
